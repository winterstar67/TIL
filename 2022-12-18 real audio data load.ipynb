### 튜토리얼이 아닌 실제 본인이 원하는 음성 데이터를 파이썬에 올리기

필자의 경우는 사이퍼즈라는 게임을 즐겨하는데, 이 게임에서 클리브라는 캐릭터의 음성파일을 받아 데이터로써 파이썬 위에 올려보도록 하겠다.

`-` 먼저 음성파일을 홈페이지에서 다운받는다.
- https://cyphers.nexon.com/game/character/voice/clive

`-` 다운을 받고 압축을 풀면, mp3 파일로 음성데이터들이 있다.

```python
import librosa
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

import torch
import torchaudio

import os

import IPython.display as ipd

import librosa.display
```

`-` 파일 path를 통해서 오디오 파일 load

```python
file = "./Clive/v_clive_awe.mp3"
```

```python
sig, sr = librosa.load(file, sr=44100)
```

```python
print(sig,sig.shape)
```

`-` **이렇게 mp3로 load를 했을 때, 필자의 경우는 데이터가 나오지 않았다.**

`-` wav의 경우는 튜토리얼에서 잘 작동을 하였기에, mp3 파일을 wav로 바꿔서 한다.  
wav를 mp3로 바꾸는 작업의 경우, 온라인 변환기나 변환 프로그램이 있어 이를 활용하여 바꾸었다.

`-` wav 파일을 만든후 Clive_wav라는 폴더에 저장하였다.

```python
file = "./Clive_wav/v_clive_awe.wav"
```

```python
sig, sr = librosa.load(file, sr=44100)
```

```python
print(sig,sig.shape)
```

파일이 잘 load가 되었다.

`-` 오디오 파일 waveplot과 실제 음성

```python
plt.figure(figsize=(6,8))
librosa.display.waveshow(sig, sr, alpha=0.5)
plt.xlabel("Time (s)")
plt.ylabel("Amplitude")
plt.title("Waveform")
```
